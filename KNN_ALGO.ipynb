```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

Load data

```python
data =load_iris()
```

```python
data
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['target'] = data.target
```

```python
df.info()
```

```python
df.head()
```

```python
x = df.drop('target',axis=1)
```

```python
y = df['target']
```

```python
x
```

```python
y
```

Split Data

```python
from sklearn.model_selection import train_test_split,GridSearchCV
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
param_grid = {
    'n_neighbors': np.arange(1, 21),   # k values from 1 to 20
    'weights': ['uniform', 'distance'], # uniform = equal weight, distance = closer points have more weight
    'metric': ['euclidean', 'manhattan', 'minkowski'] # different distance metrics
}
```

```python
grid = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
```

```python
grid.fit(x_train,y_train)
```

```python
print("best score:",grid.best_score_)
```

```python
print("best parameter:",grid.best_params_)
```

Evaluate on best test

```python
best_knn = grid.best_estimator_
```

```python
pred_y = best_knn.predict(x_test)
```

```python
pred_y
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("accuracy score:",accuracy_score(y_test,pred_y))
```

```python
print("\nClassification Report:\n", classification_report(y_test, pred_y))
```

```python
print("confusion metrix",confusion_matrix(y_test,pred_y))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn, x, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean CV score:", scores.mean())
```
